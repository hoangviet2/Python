```python
import pandas
```

### Nguyen thi mink khai

```python
dataFrameMK = pandas.read_csv("ntmk.csv")
```

```python
dataFrameMK["Họ"] = dataFrameMK["Họ"].str.upper()
dataFrameMK["tên"] = dataFrameMK["tên"].str.upper()
dataFrameMK
```

```python
MinhDucMK = dataFrameMK.groupby("Tên trường THCS").get_group("THCS MINH ĐỨC")
MinhDucMK
```
