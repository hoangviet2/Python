### link: http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/

```python
import requests
from bs4 import BeautifulSoup
```

```python
request = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
housesContent = request.content
```

```python
housesReadable = BeautifulSoup(housesContent,"html.parser")
```

```python
houses = housesReadable.find_all("div",{"class","propertyRow"})
houses[0].find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")
```

```python
l = []
for item in houses:
    data = {}
    data["Price"] = item.find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")
    data["Locality"] = item.find_all("span",{"class":"propAddressCollapse"})[1].text
    data["Address"] = item.find_all("span",{"class":"propAddressCollapse"})[0].text
    try:
        data["Beds"] = item.find("span",{"class":"infoBed"}).find("b").text
    except:
        data["Beds"] = None
    try:
        data["Area"] = item.find("span",{"class":"infoSqFt"}).find("b").text
    except:
        data["Area"] = None
    try:
        data["Bath"] = item.find("span",{"class":"infoValueFullBath"}).find("b").text
    except:
        data["Bath"] = None
    try:
        data["Halfbath"] = item.find("span",{"class":"infoValueHalfBath"}).find("b").text
    except:
        data["Halfbath"] = None
    for column_gr in item.find_all("div",{"class":"columnGroup"}):
        #print(column_gr)
        for featureGr,featureNa in zip(column_gr.find_all("span",{"class":"featureGroup"}),column_gr.find_all("span",{"class":"featureName"})):
            if "Lot size" in featureGr.text:
                data["Lot size"] = featureNa.text
    l.append(data)
```

```python
import pandas
datas = pandas.DataFrame(l)
datas
```
