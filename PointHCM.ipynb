```python
import pandas
import matplotlib.pyplot as plt
data = pandas.read_excel("Diem-XetTuyen- TPHCM-2021.xlsx")
```

### Average of point with MonChuyen

```python
chuyen = data.groupby(['MonChuyen']).mean()
plt.figure(figsize=(25,15))
plt.bar(chuyen.index,chuyen['TongDiemChuyenHoacTichHopNV12'])
print(chuyen['TongDiemChuyenHoacTichHopNV12'].max())
```

### Average Math English and Lit.

```python
Tin = data.groupby(['MonChuyen'])
chuyen = Tin.get_group('CTIN')
point = chuyen.groupby(['TongDiemChuyenHoacTichHopNV12']).count()
point = point.query('TongDiemChuyenHoacTichHopNV12 > 40')
point.query('TongDiemChuyenHoacTichHopNV12 < 44.5 & TongDiemChuyenHoacTichHopNV12 > 43.3')['SBD'].sum()
```
